```python
import cv2 
img = cv2.imread("8-bit-slice-3.jpg", cv2.IMREAD_GRAYSCALE)
print(img)
```

```python
plt.imshow(img, cmap=cm.gray)
plt.show()
```

```python
def extract_plane(pixel, n):
    n = 8-n
    d = 2**n
    return (pixel//d)%2
```

```python
n = 1
img_vector = np.vectorize(extract_plane)
trans_img = img_vector(img, n)
plt.imshow(trans_img, cmap=cm.gray)
plt.show()
```

```python
n = 5
img_vector = np.vectorize(extract_plane)
trans_img = img_vector(img, n)
plt.imshow(trans_img, cmap=cm.gray)
plt.show()
```
